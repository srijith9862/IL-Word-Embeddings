```
%%shell 
git clone https://github.com/aforakhilesh/CL-II-Project.git
git clone https://github.com/facebookresearch/fastText.git
cd fastText

cd /content/fastText
pip install .
import fasttext.util
fasttext.util.download_model('te', if_exists='ignore')
ft = fasttext.load_model('cc.te.300.bin')
import pickle
antonyms=pickle.load(open('./CL-II-Project/Gold Data/Telugu/Pickletelugu_antonyms.p', 'rb'))
plural_telugu=pickle.load(open('./CL-II-Project/Gold Data/Telugu/Pickletelugu_plural.p', 'rb'))
relationship=pickle.load(open('./CL-II-Project/Gold Data/Telugu/Pickletelugu_relationship.p', 'rb'))
verbgender=pickle.load(open('./CL-II-Project/Gold Data/Telugu/Pickletelugu_verb-gender.p', 'rb'))

antonymsvalue=0
for x in antonyms:
    a=x[0]
    b=x[1]
    c=x[2]
    d=x[3]
    d_prime = ft.get_analogies(b, a, c, k=1, on_unicode_error='strict')[0][1]
    if (d == d_prime):
      antonymsvalue = antonymsvalue+1
    print("\t- LOGIC:   " + a + ":" + b + " :: " + c + ": " + d)
    print("\t- ANALOGY: " + a + ":" + b + " :: " + c + ": " + d_prime)
    print()
print(antonymsvalue)
pluralvalue=0
for x in plural_telugu:
    a=x[0]
    b=x[1]
    c=x[2]
    d=x[3]
    d_prime = ft.get_analogies(b, a, c, k=1, on_unicode_error='strict')[0][1]
    if (d==d_prime):
      pluralvalue = pluralvalue+1
    print("\t- LOGIC:   " + a + ":" + b + " :: " + c + ": " + d)
    print("\t- ANALOGY: " + a + ":" + b + " :: " + c + ": " + d_prime)
    print()
print (pluralvalue)
relationshipvalue=0
for x in relationship:
    a=x[0]
    b=x[1]
    c=x[2]
    d=x[3]
    d_prime = ft.get_analogies(b, a, c, k=1, on_unicode_error='strict')[0][1]
    if (d==d_prime):
      relationshipvalue=relationshipvalue+1
    print("\t- LOGIC:   " + a + ":" + b + " :: " + c + ": " + d)
    print("\t- ANALOGY: " + a + ":" + b + " :: " + c + ": " + d_prime)
    print()
print (relationshipvalue)

verbgendervalue = 0
for x in verbgender:
    a=x[0]
    b=x[1]
    c=x[2]
    d=x[3]
    d_prime = ft.get_analogies(b, a, c, k=1, on_unicode_error='strict')[0][1]
    if (d==d_prime):
      verbgendervalue=verbgendervalue+1
    print("\t- LOGIC:   " + a + ":" + b + " :: " + c + ": " + d)
    print("\t- ANALOGY: " + a + ":" + b + " :: " + c + ": " + d_prime)
    print()
print (verbgendervalue)
```
